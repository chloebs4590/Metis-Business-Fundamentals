# Expanding NYC's Curbside Composting Program
***
The goal of this project is to help New York City's Department of Sanitation adopt a more comprehensive strategy for the expansion of its curbside composting program into additional community districts.

One additional it should consider is the average monthly tonnage of refuse collected per 100,000 residents in a given community district.

The below maps show the average monthly tonnage of refuse collected per 100,000 residents in the 52 community districts not currently being serviced by the curbside composting program. The time period is June 2020 - January 2022, the time period during which the program was suspended in these districts.

These two maps are interesting because they illustrate how the southern most community district on Staten Island is an outlier; when we remove it from the map, we see that approximately 3.5 times as much refuse was collected there per month, on average, during this 20-month-long period compared to the community district with the second greatest amount of refuse collected.

Based on average monthly tonnage of refuse collection alone, I would recommend that DSNY prioritize expansion into Staten Island and then eastern Queens or south central Brooklyn.
